```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X= b * np.sin(phi)
Y= a * np.cos(phi)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 

X = scaler.fit_transform(X.reshape(-1, 1)) 
Y = scaler.fit_transform(Y.reshape(-1, 1)) 
```

```python
classifier = SGDRegressor(alpha=0.001, eta0=0.001, learning_rate='constant',random_state=0)
classifier.fit(X, Y)
print(classifier.coef_, classifier.intercept_)
```

```python
reg_alpha = [0.0001, 1, 100] 
outliers = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
k = 1
plt.figure(figsize=(20,15))
for i in reg_alpha:
    for j in outliers: 
        classifier = SGDRegressor(alpha= i , eta0=0.001, learning_rate='constant',random_state=0, penalty = 'l1')
        
        classifier.fit((np.append(X, j[0])).reshape(-1, 1), (np.append(Y, j[1])).reshape(-1, 1))
        pred = classifier.predict((np.append(X, j[0])).reshape(-1, 1))
        plt.subplot(3,5,k)
#         print(classifier.coef_, classifier.intercept_)
        plt.scatter((np.append(X, j[0])).reshape(-1, 1), (np.append(Y, j[1])).reshape(-1, 1))
        plt.plot((np.append(X, j[0])).reshape(-1, 1),pred,color='red')
        plt.title(f'alpha= {i}')
        k += 1

plt.suptitle('Best-fit Linear Regression line with L1 penalty')
```

```python
reg_alpha = [0.0001, 1, 100] 
outliers = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
k = 1
plt.figure(figsize=(20,15))
for i in reg_alpha:
    for j in outliers: 
        classifier = SGDRegressor(alpha= i , eta0=0.001, learning_rate='constant',random_state=0, penalty = 'l2')
        
        classifier.fit((np.append(X, j[0])).reshape(-1, 1), (np.append(Y, j[1])).reshape(-1, 1))
        pred = classifier.predict((np.append(X, j[0])).reshape(-1, 1))
        plt.subplot(3,5,k)
#         print(classifier.coef_, classifier.intercept_)
        plt.scatter((np.append(X, j[0])).reshape(-1, 1), (np.append(Y, j[1])).reshape(-1, 1))
        plt.plot((np.append(X, j[0])).reshape(-1, 1),pred,color='red')
        plt.title(f'alpha= {i}')
        k += 1

plt.suptitle('Best-fit Linear Regression line with L2 penalty')
```
